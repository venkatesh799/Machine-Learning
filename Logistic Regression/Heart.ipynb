<h1> Hear Disease vs Chest pain types </h1>

Chest pain:

    1: Atypical angina - not related to heart
    2: Non-anginal - not related to heart
    3: Asymptomatic - not showing signs of disease

```python
pd.crosstab(data.target, data.cp)
```

```python
pd.crosstab(data.cp, data.target).plot(kind = "bar", color = ["green", "red"])
plt.title("Heart disease frequency for different chest pain")
plt.xlabel("Chest pain type")
plt.ylabel("Frequency")
plt.legend(["No Disease","Disease"]);
```

## Modeling

```python
x = data.drop('target', axis = 1)
y = data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = None)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
