```python
charges_data = data.groupby("region")["charges"].sum().sort_values(ascending = False).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x = "region", y = "charges", data = charges_data.head(), palette = 'viridis')
plt.title('Total Charges by Region')
plt.xlabel('Total Charges')
plt.ylabel('Region')
plt.show()
```

```python
charges_data = data.groupby('region')['charges'].sum().sort_values(ascending=False).reset_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='charges', y='region', data=charges_data.head(), palette='viridis')
plt.title('Total Charges by Region')
plt.xlabel('Total Charges')
plt.ylabel('Region')
plt.show()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder, StandardScaler

encoder = LabelEncoder()
```

```python
data["age"] = encoder.fit_transform(data.age)
data["sex"] = encoder.fit_transform(data.sex)
data["region"] = encoder.fit_transform(data.region)
```

```python
data["smoker"] = encoder.fit_transform(data.smoker)
```

```python
data.head()
```

```python
scaler = StandardScaler()
```

```python
standard_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

```python
standard_data.head()
```

```python
plt.figure(figsize = (12, 6))
sns.heatmap(data = standard_data.corr(), annot=True, cmap = 'coolwarm')
plt.show()
```

```python
Our Heatmap clearly says that there is high corerelation between charges and smoker
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(standard_data.drop("charges", axis = 1), standard_data["charges"], test_size = 0.25, random_state = 42)
```

```python
print(x_train.size)
x_train.head()
```

```python
print(y_train.size)
y_train.head()
```

```python
print(x_test.size)
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.score(x_test, y_test)
```
